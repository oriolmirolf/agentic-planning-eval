```python
Prompting / single‑planner variants

base — direct plan.

cot — Chain‑of‑Thought (think → output plan).

cot_sc — CoT self‑consistency: sample K CoT plans and pick the best by a judge LLM (no VAL in purple).

ltm — Least‑to‑Most (subgoals → plan).

Agentic (single run inside purple, no human/green multi‑turn)

react — Thought→Action→“ok” loop internally for T steps; finish with the plan.

tot — Tree/Beam‑of‑Thought: explore a branching tree of partial plans; a judge scores candidates; return best.

debate — two (or more) planners produce competing plans; a judge selects winner.

verifier — planner proposes; a verifier LLM checks preconditions/effects and returns corrected plan if needed.

ensemble — generate N diverse candidates (temperatures/prompts), then synthesize a single consolidated plan with a synthesizer LLM (optionally judged).
```
